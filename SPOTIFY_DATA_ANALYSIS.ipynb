# DATA ANALYSIS


## Top 5 popular artist 

```python
top_five_artist = df_music.groupby('artist').count().sort_values(by='song_title', ascending= False)['song_title'][:5]
top_five_artist
```

```python
top_five_artist.plot.barh()
plt.show
```

### Top five loudest songs

```python
top_five_loudest_tracks = df_music[['loudness', 'song_title']].sort_values(by= 'loudness', ascending= True)[:5]
top_five_loudest_tracks
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x="loudness", y="song_title", data=top_five_loudest_tracks)
plt.title("Top 5 loudes tracks")
plt.show()
```

Artist with the most danceability song 

```python
top_five_artists_danceability_songs = df_music[['danceability', 'song_title', 'artist']].sort_values(by='danceability', ascending= False)[:5]
top_five_artists_danceability_songs
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='danceability', y= 'artist', data=top_five_artists_danceability_songs)
plt.title('Artist with most danceability Songs')
plt.show
```

Top 10 Instrumental songs


```python
top_ten_instrumental_tracks = df_music[['instrumentalness', 'song_title', 'artist']].sort_values(by='instrumentalness', ascending=False)[:5]
top_ten_instrumental_tracks
```

```python
plt.figure(figsize=(12, 7))
plt.pie(x='instrumentalness', data=top_ten_instrumental_tracks, autopct='%1.2f%%', labels=top_ten_instrumental_tracks.song_title)
plt.show
```

```python
interest_feature_cols = ['tempo', 'loudness', 'acousticness', 'danceability', 'duration_ms', 'energy', 'instrumentalness', 'liveness', 'speechiness', 'valence']   
```

```python
for feature_col in interest_feature_cols:
    pos_data = df_music[df_music['target'] == 1 ][feature_col]
    neg_data = df_music[df_music['target'] == 0 ][feature_col]

    plt.figure(figsize= (12, 7))

    sns.distplot(pos_data, bins=30, label='Positive', color= 'green')
    sns.distplot(neg_data, bins=30, label='Negative', color= 'red')

    plt.legend(loc='upper right')
    plt.title(f'Positive to Negative Histogram Plot For {feature_col}')
    plt.show
```
